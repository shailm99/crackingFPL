```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 
import json
```

```python
path='players_1920_fin.csv'
fpl20=pd.read_csv(path,parse_dates=['kickoff_time'])
player_df=fpl20.copy()
player_df.head()
```

```python
player_df.columns
```

```python
player_df.drop(columns=['Unnamed: 0','assists', 'bonus', 'clean_sheets','goals_conceded', 'goals_scored','own_goals',
       'penalties_missed', 'penalties_saved', 'red_cards', 'saves',
       'transfers_balance', 'transfers_in', 'transfers_out', 'yellow_cards'],inplace=True)
```

# Features I would like engineer 
* 10 GW MA total points 
* 5 GW MA Total Points
* Last GW MA Total Points 

```python
d=dict(player_df.dtypes)
for col,dtype in d.items():
    if dtype==np.int64 or dtype==np.float64:
        sns.distplot(player_df[col])
        plt.show()
```

```python
player_df.sort_values(by=['full','kickoff_time'],inplace=True)
player_df.set_index('full',inplace=True)
player_df['Time of Kickoff']=player_df['kickoff_time'].dt.time
player_df['kickoff_day']=player_df['kickoff_time'].dt.strftime('%Y-%m-%d')
```

```python
for i in ['total_points', 'ict_index', 'bps', 'threat', 'influence', 'creativity']:
    player_df['{} {} Week MA'.format(i,'5')]=None
    player_df['{} {} Week MA'.format(i,'1')]=None
    player_df['{} {} Week MA'.format(i,'10')]=None

def MA(time,player,attribute):
    df=player_df[['kickoff_day',attribute]].loc[player]
    series=pd.Series(df[attribute],index=df.index)
    MA=(series.rolling(time+1).sum() - series)/time
    lst=list(MA)
    integ=[round(float(i),2) for i in lst]
    player_df['{} {} Week MA'.format(attribute,time)].loc[player]=integ
    return player_df['{} {} Week MA'.format(attribute,time)].loc[i]
```

```python
player_names=list(player_df.index.unique())
for i in player_names:
    for atr in ['total_points', 'bps', 'ict_index', 'threat', 'influence', 'creativity']:
        for t in [1,5, 10]:
            MA(time=t,player=i,attribute=atr)
```

```python
player_df.drop(columns=['kickoff_time','ict_index','bps','team_a_score','team_h_score', 'threat', 'influence', 'creativity'],inplace=True)
```

```python
player_df.sort_values(by='kickoff_day',inplace=True)
player_df.dropna(inplace=True)
player_df.reset_index(inplace=True)
```

```python
player_df = player_df[player_df['minutes'] > 60]
```

```python
for i in ['total_points 5 Week MA', 'total_points 1 Week MA', 'total_points 10 Week MA', 
         'ict_index 5 Week MA', 'ict_index 1 Week MA', 'ict_index 10 Week MA',
       'bps 5 Week MA', 'bps 1 Week MA', 'bps 10 Week MA']:
    player_df[i]= player_df[i].astype('float')
```

```python
player_df['sd'] = player_df['total_points']
home = player_df[['was_home', 'total_points', 'fixture' ,  'sd']]
home = home.groupby('was_home')
home = home.agg({'total_points': 'mean' , 'fixture' : 'count', 'sd' : 'std'}).reset_index()
home.rename(columns  = {'fixture' : 'number of fixtures', 'total_points' : 'average points'})
```

```python
sns.barplot(home['was_home'], home['total_points'])
plt.xlabel('Played at Home?')
plt.ylabel('Average Points Per Game')
plt.title('Does Playing at Home Impact a Player\'s performance')
```
